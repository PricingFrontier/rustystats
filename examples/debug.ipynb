```python
import polars as pl
import rustystats as rs

data = pl.read_parquet(
    "https://raw.githubusercontent.com/PricingFrontier/pricing-data-example/"
    "917c853e256df8d5814721ab56f72889a908bb08/data/processed/frequency_set.parquet"
)
train_data = data.filter(pl.col("Group") <= "3")

glm_model = rs.glm_dict(
    response="ClaimCount",
    terms={
        "BonusMalus": {"type": "ns"},
        "VehAge": {"type": "linear"},
        "DrivAge": {"type": "linear"},
        "Region": {"type": "target_encoding"},
        "VehBrand": {"type": "target_encoding"},
    },
    interactions=[
        {
            "VehAge": {"type": "linear"},
            "BonusMalus": {"type": "ns"},
            "include_main": False
        }
    ],
    data=train_data,
    family="poisson",
    offset="Exposure",
)

result = glm_model.fit()
result.to_bytes()  # Does this fail?
```
