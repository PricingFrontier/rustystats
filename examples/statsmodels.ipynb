```python
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf

data = pd.read_parquet("https://raw.githubusercontent.com/PricingFrontier/pricing-data-example/917c853e256df8d5814721ab56f72889a908bb08/data/processed/frequency_set.parquet")
```

```python
model = smf.glm(
    formula="ClaimCount ~ VehPower + VehAge + C(Area) + C(Region)",
    data=data,
    family=sm.families.Poisson()
).fit()
```

```python
model.summary(alpha=0.05)
```
